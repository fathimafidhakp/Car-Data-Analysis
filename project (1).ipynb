### Conclusion

The car dataset shows how different factors affect the selling price of cars. The most important factors are **year, kilometers driven, engine power, and ownership**. Newer cars usually have higher prices because they are in better condition and have more modern features. Cars with lower kilometers driven are also more valuable because they are less used.

Engine capacity and maximum power also play a big role. Cars with higher power and bigger engines generally have higher prices. Fuel type also has some influence. Diesel cars often have slightly higher value because they give better mileage and are preferred for long-distance travel.

Most of the cars in the dataset use **manual transmission**, and many are sold by **individual owners** instead of dealers. Cars with **first ownership** have better value compared to second or third owners, as they are considered more reliable.

The **torque column was removed** from the dataset because it had inconsistent or complex values, making it difficult to analyze properly.

Overall, the dataset clearly shows that **car condition, performance, and usage history** are the main factors that determine the price. Buyers usually prefer cars that are newer, less used, and more powerful.
